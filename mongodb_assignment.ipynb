```python
import pandas as pd
import pymongo
import json
```

```python
csv_file=r"https://raw.githubusercontent.com/vigneshk/Admission-Dataset/master/Admission.csv"
```

```python
admission_details1 = pd.read_csv(csv_file)
```

```python
json_output = r"admission_details1.json"
output = admission_details1.to_json(json_output,indent=1,orient='records')
```

```python
output1 = pd.read_json("admission_details1.json")
```

```python

records = json.loads(output1.to_json(orient='records'))
```

```python
type(records)
```

```python
client = pymongo.MongoClient("mongodb://localhost:27017/")
```

```python
dataBase = client["Ineuron"]
```

```python
collection = dataBase["Addmision_details"]
```

```python
collection.insert_many(records)
```

```python
all_record = collection.find({})
```

```python
for i in all_record:
    print(i)
```

```python
result_gre = collection.find({},{'GRE Score','TOEFL Score'}).limit(5)
```

```python
for i in result_gre:
    print(i)
```

```python
my_query1={"GRE Score":"314"}
```

```python
result_GRE=collection.find(my_query1)
```

```python
for i in result_GRE:
    print(i)
```

```python
my_query2={"GRE Score":{"$gt":"337"}}
```

```python
result_gt=collection.find(my_query2)
```

```python
for row in result_gt:
    print(row)
```
